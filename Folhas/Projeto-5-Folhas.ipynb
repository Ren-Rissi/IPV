```r
library(ggplot2)
library("jpeg")
library(grid)
library("ggfortify")
library("stats")

img<- readJPEG("f10.jpg")
dim = dim(img)
imatriz <- matrix(img,dim[1], dim[2])

tmpmtx = matrix(0,dim[1],dim[2])


#limiarização 

for(i in 1:dim[1]){
    for(j in 1:dim[2]){
        if(imatriz[i,j] >= 0.32){
            tmpmtx[i,j] = imatriz[i,j]
            
        }
    }
}

#image(tmpmtx, col = gray((0:256)/256))
```
